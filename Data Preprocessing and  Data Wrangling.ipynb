```python
train.head()
```

```python
train['Average_Cost'].apply(type)
```

```python
train['Average_Cost'] =train.Average_Cost.replace('[₹,]', '', regex =True)
test['Average_Cost'] =test.Average_Cost.replace('[₹,]', '', regex =True)
```

```python
train['Minimum_Order'] =train.Minimum_Order.replace('[₹]', '', regex =True)
test['Minimum_Order'] =test.Minimum_Order.replace('[₹]', '', regex =True)
```

```python
train['Rating'] =train.Rating.replace('[- NEW Opening Soon Temporarily Closed]', -1, regex =True)
test['Rating'] =test.Rating.replace('[- NEW Opening Soon Temporarily Closed]', -1, regex =True)
```

```python
train['Delivery_Time'] =train.Delivery_Time.replace('[ minutes]', '', regex =True)
```

```python
train['Votes'] =train.Votes.replace('[-]', -999, regex =True)
test['Votes'] =test.Votes.replace('[-]', -999, regex =True)
```

```python
train['Reviews'] =train.Reviews.replace('[-]', -999, regex =True)
test['Reviews'] =test.Reviews.replace('[-]', -999, regex =True)
```

```python
train['Reviews']=train['Reviews'].astype(int)
test['Reviews']=test['Reviews'].astype(int)

train['Votes']=train['Votes'].astype(int)
test['Votes']=test['Votes'].astype(int)

train['Rating']=train['Rating'].astype(float)
test['Rating']=test['Rating'].astype(float)

train['Minimum_Order']=train['Minimum_Order'].astype(int)
test['Minimum_Order']=test['Minimum_Order'].astype(int)

train['Delivery_Time']=train['Delivery_Time'].astype(int)
```
